```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

N = 1000
n = 20
```

### Task 1

```python
plt.plot(range(N), sorted([stats.kstest(stats.norm.rvs(size=n), cdf=stats.norm.cdf)[1] for _ in range(N)]))
plt.show()
```

### Task 2

```python
y = np.zeros(N)
for i in range(N):
    x = stats.norm.rvs(size=n)
    mn = np.mean(x)
    u = stats.norm.cdf(sorted(x), loc=mn)
    y[i] = stats.kstest(x, cdf=lambda _: u)[1]

plt.plot(range(N), sorted(y))
plt.show()
```
