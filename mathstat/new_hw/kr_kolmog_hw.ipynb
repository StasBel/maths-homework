```python
%matplotlib inline
import numpy as np
from scipy import stats as sc
import matplotlib.pyplot as plt
n = 1000
k = 20
```

## task1

```python
plt.plot(range(n), sorted(sc.kstest(sc.norm.rvs(size=k), sc.norm.cdf)[1] for _ in range(n)))
```

## task2

```python
def gen():
    x = sc.norm.rvs(size=k)
    return sc.kstest(x, lambda _: sc.norm.cdf(sorted(x), loc=x.mean()))[1]

plt.plot(range(n), sorted(gen() for _ in range(n)))
```
