```r
N = 1000
n = 20
```

## task1

```r
x = as.integer(seq(1, N, length=N))
y = unlist(lapply(x, function(z) {ks.test(rnorm(n), "pnorm")$p.value}))
plot(x, sort(y), col="blue", type="l")
```

```r
x = as.integer(seq(1, N, length=N))
gen = function(z) {
    r = rnorm(n)
    cdf = pnorm(sort(r), mean=mean(r))
    return(ks.test(r, cdf, alternative="two.sided")$p.value)
}
y = unlist(lapply(x, gen))
plot(x, sort(y), col="blue", type="l")
```
