## read data

```r
library(foreign)
data <- data.matrix(read.dta("deaths.dta"))
data <- scale(data)
data
```

## test train split

```r
ratio <- 0.2
n <- nrow(data)
sample <- sample.int(n = nrow(data), size = floor(ratio * n), replace = F)
data_test <- data[sample, ]
data_train <- data[-sample, ]
```

## depencies

```r
colsx = c("teachers", "pop", "year")
colsy = c("deaths")
```

## linear regression

```r
x <- data_train[,colsx]
A <- data_train[,colsx]
A <- cbind(1, A)
At <- t(A)
y <- data_train[,colsy]
w <- solve(At %*% A) %*% At %*% y
```

## test

```r
x <- data_test[,colsx]
y <- data_test[,colsy]
y_pred <- w[1] + w[2] * x
SSE <- sum((y_pred - y) ^ 2)
SSE
```
